```python
def calcular_media(nota1,nota2,nota3):
    media = (nota1 + nota2 + nota3)/3
    return media
try:
    n1 = float(input("Digite: "))
    n2 = float(input("Digite: "))
    n3 = float(input("Digite: "))
    
    media_final = calcular_media(n1,n2,n3)
    if media_final >= 7.0 :
        print("Aprovado")
    else:
        print("Reprovado")
except ValueError:
    print("Digite um número válido.")
```

```python
def eh_primo(numero):
    
    if numero < 2:
        return False
    elif numero ==2 or numero ==3:
        return True
    elif numero % 2 == 0 or numero % 3 == 0:
        return False
    
    
    for i in range(3,int(numero**0.5)+1,2):
        if numero % i == 0:
            return False
        else:
            return True
        
try:
    num = int(input("Digite o número: "))
    
    if eh_primo(num):
        print("Número ",num," é primo")
    else:
        print("Número ",num," não é primo")
        
except ValueError:
    print("Digite apenas números válidos.")
```

```python
def celsius_farenheit(celsius):
    farenheit = celsius * 1.8 + 32
    return farenheit
try:
    temp_celsius = float(input("Celsius: "))
    temp_farenheit = celsius_farenheit(temp_celsius)
    
    print(f"Celsius equivale à : {temp_farenheit}")
    
except ValueError:
    print("Digite apenas números. ")
```

```python
def par_impar(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Impar"
    
try:
        num = int(input("Digite o número: "))
        
        resultado = par_impar(num)
        
        print({resultado})
        
        
except ValueError:
        print("Digite um número válido. ")
```
