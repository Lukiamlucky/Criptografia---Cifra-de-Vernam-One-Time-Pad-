```python
alunos = {} # Dicionário global para armazenar os dados

def cadastrar_alunos(alunos_dict):
    """Permite ao usuário cadastrar nomes e notas de alunos."""
    print("\n--- Opção 1: Cadastrar Alunos ---")
    try:
        qtd = int(input("Digite a quantidade de alunos a cadastrar: "))
        
        for entrada in range(qtd):
            nome_aluno = input(f"Digite o nome do {entrada + 1}º aluno: ")
            try:
                nota_aluno = float(input(f"Digite a nota do aluno {nome_aluno}: "))
                alunos_dict[nome_aluno] = nota_aluno
            except ValueError:
                print("ERRO: Nota inválida. Este aluno não foi cadastrado.")
        
        print(f"\nAlunos cadastrados: {alunos_dict}")
        
    except ValueError:
        print("ERRO: Quantidade inválida.")

def listar_alunos(alunos_dict):
    """Lista todos os nomes dos alunos cadastrados."""
    print("\n--- Opção 2: Lista de Alunos ---")
    if not alunos_dict:
        print("Nenhum aluno cadastrado.")
        return
        
    for nome in alunos_dict.keys():
        print(f"- {nome}")

def calcular_media(alunos_dict):
    """Calcula e exibe a média das notas dos alunos."""
    print("\n--- Opção 3: Média das Notas ---")
    if not alunos_dict:
        print("Nenhum aluno cadastrado para calcular a média.")
        return
        
    notas = alunos_dict.values()
    media = sum(notas) / len(notas) # Uso da função sum() para simplificar
    
    print(f"A média das notas é: {media:.2f}")

def pesquisar_aluno(alunos_dict):
    """Permite pesquisar a nota de um aluno específico."""
    print("\n--- Opção 4: Pesquisar Aluno ---")
    if not alunos_dict:
        print("Nenhum aluno cadastrado para pesquisa.")
        return
        
    pesquisa = input("Digite o nome a ser pesquisado: ")
    
    if pesquisa not in alunos_dict:
        print(f"Não existe '{pesquisa}' no dicionário.")
    else:
        print(f"A nota do aluno '{pesquisa}' é: {alunos_dict[pesquisa]:.2f}")

def main():
    """Função principal que executa o loop do menu."""
    
    # O dicionário 'alunos' será usado como um objeto global/compartilhado
    global alunos 
    
    while True:
        # --- Menu ---
        print("\n--- Sistema de Gerenciamento de Notas ---")
        print("1 - Cadastrar Alunos e Notas")
        print("2 - Listar Nomes dos Alunos")
        print("3 - Calcular e Mostrar Média das Notas")
        print("4 - Pesquisar Nota de um Aluno")
        print("5 - Sair do Programa")
        print("-----------------------------------------")
        
        opcao = input("Digite uma opção de 1 a 5: ")

        if opcao == '1':
            cadastrar_alunos(alunos)
        
        elif opcao == "2":
            listar_alunos(alunos)
            
        elif opcao == "3":
            calcular_media(alunos)

        elif opcao == "4":
            pesquisar_aluno(alunos)
            
        elif opcao == "5":
            print("Encerrando programa. Até mais!\n")
            break  # Sai do loop
        
        else:
            print("Opção inválida. Digite um número de 1 a 5.")

# A linha abaixo garante que a função main() seja chamada ao executar o script.
if __name__ == "__main__":
    main()
```
