```python
# ===============================
# GERENCIADOR DE ALUNOS - 
# ===============================

alunos = []  # lista vazia para armazenar os nomes

# 1. EXIBIÇÃO DO MENU (Sem Loop)
print("\n===== MENU PRINCIPAL =====")
print("1 - Adicionar aluno")
print("2 - Mostrar lista de alunos")
print("3 - Remover aluno (transferido)")
print("4 - Ordenar lista de alunos")
print("5 - Contar alunos cadastrados")
print("6 - Limpar toda a lista")
print("7 - Sair")

# 2. ESCOLHA DA OPÇÃO
opcao = input("Escolha uma opção (1-7): ")

# 3. EXECUÇÃO DA OPÇÃO (O programa termina após este bloco)

# 1️⃣ Adicionar aluno
if opcao == "1":
    nome = input("Digite o nome do aluno: ").strip().title()
    if nome:
        alunos.append(nome)
        print(f"Aluno '{nome}' adicionado com sucesso!")
        
        # Para demonstrar que a inserção funcionou, mostramos a lista
        print("\n--- LISTA APÓS AÇÃO ---")
        for i, nome in enumerate(alunos):
            print(f"{i+1}. {nome}")

# 2️⃣ Mostrar lista de alunos
elif opcao == "2":
    print("\n--- LISTA DE ALUNOS ---")
    if not alunos:
        print("Nenhum aluno cadastrado.")
    else:
        for i, nome in enumerate(alunos):
            print(f"{i+1}. {nome}")

# 3️⃣ Remover aluno (transferido)
elif opcao == "3":
    if not alunos:
        print("A lista está vazia!")
    else:
        # Nota: Sem um loop para pré-inserir dados, esta opção só funcionará se você rodar o código com dados pré-definidos.
        print("Opção 3 executada. Funcionalidade de remoção aqui.")

# 4️⃣ Ordenar lista de alunos
elif opcao == "4":
    if not alunos:
        print("Nenhum aluno para ordenar.")
    else:
        print("Opção 4 executada. Funcionalidade de ordenação aqui.")

# 5️⃣ Contar quantos alunos existem
elif opcao == "5":
    print(f"Total de alunos cadastrados: {len(alunos)}")

# 6️⃣ Limpar a lista completamente
elif opcao == "6":
    confirmar = input("Tem certeza que deseja apagar toda a lista? (s/n): ")
    if confirmar.lower() == "s":
        alunos.clear()
        print("Lista de alunos apagada com sucesso!")

# 7️⃣ Sair do programa
elif opcao == "7":
    print("Encerrando o sistema... Até logo!")

# Opção inválida
else:
    print("Opção inválida!")

print("\n--- PROGRAMA FINALIZADO ---")
```
