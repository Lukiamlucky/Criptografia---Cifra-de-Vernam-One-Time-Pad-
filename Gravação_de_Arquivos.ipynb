```python
matriz = []
print("Digite os valores para a matriz 3x3:")
    
for linha in range(3):
    linhas = []
    for coluna in range(3):
        valor = int(input(f"Elemento [{linha+1}][{coluna+1}]: "))
        linhas.append(valor)
        matriz.append(linhas)
    with open('matriz.txt', 'w', encoding='utf-8') as arquivo:
        for linha in matriz:
            linha_str = ' '.join(str(x) for x in linha)
            arquivo.write(linha_str + '\n')
```

```python
matriz = []
with open('matriz.txt', 'r', encoding='utf-8') as arquivo:
    for linha in arquivo:
        elementos = [int(x) for x in linha.strip().split()]
        matriz.append(elementos)

        soma = 0
        for linha in matriz:
            for elemento in linha:
                soma += elemento
        
        for linha in matriz:
            print(linha)
        
        print(f"Soma de todos os elementos: {soma}")
```
